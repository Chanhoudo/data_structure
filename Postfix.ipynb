```python
class Stack():
  def __init__(self):
    self.stack = []
    self.top = -1
  def isEmpty(self):
    if self.top == -1:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
    self.top += 1
  def peek(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    return self.stack[self.top]
  def pop(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    self.top -= 1
    return self.stack.pop(self.top+1)
  def size(self):
    return self.top+1
  def clear(self):
    self.stack = []
    self.top = -1
  def display(self):
    for i in range(self.top + 1):
      print(i+1, "번째 나올 항목",  self.stack[self.top - i])
```

```python
class Postfix(Stack):
  def __init__(self):
    super().__init__()
    self.cal = ['+', '-', '*', '/']
  def postfix(self, mat):
    self.clear()
    for c in mat:
      if c in self.cal:
        if self.size() < 2:
          return False
        right = self.pop()
        left  = self.pop()
        if c == '+':
          self.push(left + right)
        elif c == '-':
          self.push(left - right)
        elif c == '*':
          self.push(left * right)
        elif c == '/':
          self.push(left / right)
      else:
        self.push(int(c))
    if self.size() == 1:
      return self.stack[0]
  def in_change_pos(self, mat):
    oper = {'+' : 1, '-' : 1, '*' : 2, '/' : 2}
    self.clear()
    number = Stack()
    n = 0
    for c in mat:
      if c in self.cal:
        if self.size() > 0 and oper[self.peek()] >= oper[c]:
          number.push(self.pop())
          self.push(c)
        else:
          self.push(c)
      else:
        number.push(int(c))
    for i in range(self.size()):
      number.push(self.pop())
    return number



post = Postfix()

answer = post.postfix('82/3-32*+')
print(answer)

answer = post.in_change_pos('8/2-3+3*2')
print(answer.stack)
```
