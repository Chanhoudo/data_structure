```python
class Stack():
  def __init__(self):
    self.stack = []
    self.top = -1
  def isEmpty(self):
    if self.top == -1:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
    self.top += 1
  def peek(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    return self.stack[self.top]
  def pop(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    self.top -= 1
    return self.stack.pop(self.top+1)
  def size(self):
    return self.top+1
  def clear(self):
    self.stack = []
    self.top = -1
  def display(self):
    for i in range(self.top + 1):
      print(i+1, "번째 나올 항목",  self.stack[self.top - i])

stack = Stack()

stack.push(2)
stack.push(3)
stack.push(1)
stack.push(7)
stack.push(6)

stack.display()

print("맨 위에 있는 항목 삭제:", stack.pop())
stack.push(1)

print("맨 위에 있는 항목 :", stack.peek())

print("스택의 사이즈 :", stack.size())


stack.display()

print("비어있나요? :", stack.isEmpty())
stack.push(3)
print("비어있나요? :", stack.isEmpty())

stack.clear()
stack.display()
print("비어있나요? :", stack.isEmpty())
```

```python
def gard_check(mat):
  check = Stack()
  for m in mat:
    if m == '{' or m == '(' or m == '[':
      check.push(m)
    elif m == '}':
      if '{' != check.pop():
        print("잘못됐습니다.")
        return False
    elif m == ']':
      if '[' != check.pop():
        print("잘못됐습니다.")
        return False
    elif m == ')' :
      if '(' != check.pop():
        print("잘못됐습니다.")
        return False
  if check.size() == 0:
    print("완성됐습니다.")
    return True
  else:
    print("실패")
    return False


mat = "{A[i+1]) = 0}"
print(gard_check(mat))
print(gard_check("if((i==0) && (j==0)"))
print(gard_check("While(it<10)) {it-=1}"))
print(gard_check("A[(i+1]) = 0"))

```
